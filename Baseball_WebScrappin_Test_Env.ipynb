```python
#Scrape Batting Table
def battin_table(soup,season):

    # Buscar la tabla de bateo
    table_rows_batting = soup.find(id="teams_standard_batting").find_all("tr")

    l = []
    for tr in table_rows_batting:
        td = tr.find_all('td')
        th = tr.find_all('th')
        row = [tr.text for tr in td]
        team  = [tr.text for tr in th]
        complete_row = team + row
        l.append(complete_row)

    batting_teams_df=pd.DataFrame(l[1:-2], columns=l[0])
    
    batting_teams_df["Years"] = season
    
    return batting_teams_df

#Scrape Pitching Table
def pitching_table(soup,season):
    # Buscar la tabla de pitcheo
    table_rows_pitching = soup.find(id="teams_standard_pitching").find_all("tr")

    l = []
    for tr in table_rows_pitching:
        td = tr.find_all('td')
        th = tr.find_all('th')
        row = [tr.text for tr in td]
        team  = [tr.text for tr in th]
        complete_row = team + row
        l.append(complete_row)

    pitching_teams_df=pd.DataFrame(l[1:-2], columns=l[0])
    
    pitching_teams_df["Years"] = season
    
    return pitching_teams_df

#Scrape wins abv avg Table
def output_table(soup,season):
    # Buscar la tabla de output
    table_rows_output = soup.find(id="team_output").find_all("tr")
    
    l = []
    for tr in table_rows_output:
        td = tr.find_all('td')
        th = tr.find_all('th')
        row = [tr.text for tr in td]
        team  = [tr.text for tr in th]
        complete_row = team + row
        l.append(complete_row)

    output_teams_df=pd.DataFrame(l[1:], columns=l[0])
    output_teams_df["Years"] = season
    
    return output_teams_df

#Scrape filding table
def fielding_table(soup,season):
    # Buscar la tabla de fielding
    table_rows_fielding = soup.find(id="teams_standard_fielding").find_all("tr")

    
    l = []
    for tr in table_rows_fielding:
        td = tr.find_all('td')
        th = tr.find_all('th')
        row = [tr.text for tr in td]
        team  = [tr.text for tr in th]
        complete_row = team + row
        l.append(complete_row)

    fielding_teams_df=pd.DataFrame(l[1:-2], columns=l[0])
    fielding_teams_df["Years"] = season
    
    return fielding_teams_df

#Formula para iterar entre temporadas
def season_table_scrap(start,end):
    seasons = np.arange(start,end+1,1)
    
    
    
    # Inicializar listas para acumular los DataFrames de cada temporada
    batting_list = []
    pitching_list = []
    output_list = []
    fielding_list = []
    
            
    # Iniciar el navegador (asegúrate de tener un controlador como ChromeDriver instalado)
    driver = webdriver.Chrome()
    
    for season in seasons:
        
        # Navegar a la página
        driver.get(f"https://www.baseball-reference.com/leagues/majors/{season}.shtml")
        # Esperar a que el elemento esté presente
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "teams_standard_fielding")))

        # Obtener el HTML de la página
        page_content = driver.page_source

        # Parsear el HTML con BeautifulSoup
        soup = BeautifulSoup(page_content, "html.parser")
        
        # Extraer y acumular los DataFrames para cada tabla
        batting_list.append(battin_table(soup, season))
        pitching_list.append(pitching_table(soup, season))
        output_list.append(output_table(soup, season))
        fielding_list.append(fielding_table(soup, season))

    # Cerrar el navegador
    driver.quit()

    # Concatenar todos los DataFrames acumulados
    batting = pd.concat(batting_list, ignore_index=True)
    pitching = pd.concat(pitching_list, ignore_index=True)
    output = pd.concat(output_list, ignore_index=True)
    fielding = pd.concat(fielding_list, ignore_index=True)
        
    
    return batting, pitching, output, fielding
     
```

```python
batting, pitching, output, fielding = season_table_scrap(2010,2014)
```
